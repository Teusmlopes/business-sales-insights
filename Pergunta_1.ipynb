```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_vendas = pd.read_csv('Dados/dataset.csv')
```

```python
df_vendas['Data_Pedido'] = pd.to_datetime(df_vendas['Data_Pedido'], dayfirst=True)
df_vendas['Ano'] = df_vendas['Data_Pedido'].dt.year
```

```python
df_vendas_agrupado = df_vendas.groupby(['Ano', 'Segmento'])['Valor_Venda'].sum()
df_vendas_agrupado
```

```python
plt.figure(figsize=(10,5))
df_vendas_agrupado.unstack().plot(kind='bar', figsize=(10,5))
plt.title("Total de Vendas por Segmento e Ano")
plt.xlabel("Ano")
plt.ylabel("Total de Vendas")
plt.legend(title="Segmento")
plt.xticks(rotation=0)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
